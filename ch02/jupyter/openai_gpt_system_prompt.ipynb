# system prompt vs. user prompt

This notebook demonstrates the difference between **system prompt** (system message) and the **user prompt** when calling an LLM.

We send the *same* user request with and withoug a system message to compare the outputs.

## 1. Setup

## 2. Function to query the model

This section defines a helper function to send a user prompt to the model.

```python
def query_model(instructions: str | None, prompt: str, model: str = "gpt-4.1-mini", temperature: float = 0) -> str:
  client = OpenAI()

  messages = []
  if instructions:
      messages.append({"role": "system", "content": instructions})
  messages.append({"role": "user", "content": prompt})

  chat_completion = client.chat.completions.create(
    model=model,
    messages=messages,
    temperature=temperature,
  )

  return chat_completion.choices[0].message.content
```

## 3. Send user prompt

This section sends and user prompt to the model using the function previously defined.

```python
instructions = """You are a strict grammar teacher. Always respond in one sentence and correct any mistakes."""
prompt = "Explain me what is context engineering in simple words"
model = "gpt-4.1-mini"
temperature = 0

response = query_model(instructions, prompt, model, temperature)
print("=== With system prompt ===")
print(f"User query: {prompt}")
print(f"Response: {response}")

response = query_model(None, prompt, model, temperature)
print("=== With only user prompt ===")
print(f"User query: {prompt}")
print(f"Response: {response}")
```
